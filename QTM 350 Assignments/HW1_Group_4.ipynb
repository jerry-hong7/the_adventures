## Q2: Linux Basics
In this section, basic Linux operations will be discussed such as: 1) determining total memory available in a system; 2) the version of Linux being used; 3) updating the Linux version; and 4) file management.

### Memory
The `free` command can be used to check the available memory in the system. The `--human` option is used to format the memory in a human-readable output.

````
%%bash
free --human
````

By using the `free` command, it is shown that this Google Colab notebook has 12 Gigabytes of memory.

### Linux Version
The `uname` command is used to print basic information about the kernel, operating system, and system hardware. This command can be used to check the Linux version being run. The `-s` option prints the kernel name and the `-r` option prints the kernel release version.

````
%%bash
uname -s -r
````

From `uname`, it is shown that this Google Colab notebook is using Linux 5.15.109+.

### Update Linux
To update the system, the commands `sudo apt update` and `sudo apt upgrade` must be used. The command `sudo` stands for "Superuser Do" and it allows programs to run with admin privileges. The command `apt` stands for "Advanced Package Tool." This command is a collection of tools for managing software packages on Linux. The `update` function writes a list of packages that can be updated and the `upgrade` function installs the new packages that were listed.

````
%%bash
sudo apt update
sudo apt upgrade --yes #yes option chooses "yes" to any prompts
````

### File Management
A new file can be created with the `touch` command. However, the standard output of a program can be immediately redirected into a new file using the `>` operator. This operator will overwrite a file if it already exists. The `>>` operator can be used to append standard output to an existing file without overwriting.


#### Example of file creation and redirection
The output of `free --human` can be stored in a file using the `>` operator. The contents of SystemInfo.txt can be viewed with `cat SystemInfo.txt`.

````
%%bash
free --human > SystemInfo.txt
cat SystemInfo.txt
````

As seen by the `cat` command, SystemInfo.txt contains information about the system's memory. Additional information can be added to this file using the `>>` operator.

````
%%bash
uname -s -r >> SystemInfo.txt
cat SystemInfo.txt
````

#### File Analysis
The command `cat` is one example of a useful tool to analyze files. Other useful commands are `wc` and `cut`.

`wc` stands for "word count" and is used to count number of lines, words, and or characters. To get the number of lines, the `-l` option must be used.

````
%%bash
wc -l SystemInfo.txt
````

The value 4 represents the number of lines in SystemInfo.txt. To remove "SystemInfo.txt" from the output, the `<` operator can be used. This operator redirects SystemInfo.txt into the standard input of the `wc` command.

````
%%bash
wc -l < SystemInfo.txt
````

The command `cut` is used for selecting data from each line of a file and writing the results to standard output. This can be useful for selecting columns from delimiter-separated data. In SystemInfo.txt, the first column (`--fields 1`) can be selected by using a space as a delimeter (`--delimiter' '`).

````
%%bash
cut --delimiter ' ' --fields 1 SystemInfo.txt
````

The first word of each line was displayed. Note that only 3 values were displayed because the first line does not have a word in the first column.

I defined this function to calculate the annual rate of increase:





````
import numpy as np

def calculate_annual_rate(x1, x2, y1, y2):
    t = y1 - y2
    r = np.log(x1 / x2) / t
    return r
````

### Task 1

To determine how many years it would take the saver profile to afford a house which costs $312,196, we will need a new formula. If we manipulate the formula from example 1, we have

$$t = \log_{r+1}{((Price/HouseSpend) * r + 1)}$$

We will use the math package to calculate the logarithm and set rates at 3% for this example.

````
import math
````

In order to reuse our code for other examples, we'll write a function called `time_to_save`, which takes as parameters, the home price of interest, the interest rate, and the house_spend variable (1/4 gross income). The function will return the time, in years, that it would take for the saver profile to afford a home at the given price.

````
def time_to_save(price, rate, hs):
  rate /= 100
  return math.log(((price/hs) * rate + 1), (1+rate))
````

````
time_to_save(312196, 3, house_spend)
````
