# Human Socioeconomic Status and Morphometrics

Human morphometrics vary tremendously, and are dependent on both internal (genetic) and external (environmental) factors. At times, environmental trauma can even cause epigenetic remodelling, leading to generational changes causing drastic phenotypic change in a non-mendelian fashion.

Environmental factors regarding human size are highly dependent on diet during developmental stages, which in turn is effected by personal socioeconomic status.

We used the brfss dataset to examine relationships between environmental factors and physical phenotype in humans.

We began by importing the brfss dataset

```
from os.path import basename, exists

def download(url):
    filename = basename(url)
    if not exists(filename):
        from urllib.request import urlretrieve
        local, _ = urlretrieve(url, filename)
        print('Downloaded ' + local)

download('https://github.com/AllenDowney/' +
         'ElementsOfDataScience/raw/master/brfss.hdf5')
```

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data = pd.read_hdf('brfss.hdf5', 'brfss')
data.dropna()
data.head()
```

We then created a violinplot to visualize the probility distributions of height based on the income groups.

```
sns.violinplot(x='INCOME2', y='HTM4', data=data, inner=None, palette = 'plasma')

plt.xlabel('Income Group')
plt.ylabel('Height (cm)')
plt.title('Height vs. Income')
```

Given the figure, it seems that the higher income groups have less variance in the possible range of heights compared to lower income groups. In addition, their probability distributions are more evenly spread out whereas the lower groups are more left skewed.

However, these slight differences between height and income group are not drastic enough to make any discernable conclusions between socioeconomic status and morphometric change alone.

To look further, we created boxplots to visualize the distriubution of weight with respect to income groups. Additionally, this allowed us to the opportunity to observe any glaring outliers present in certain income groups.

```
sns.boxplot(x='INCOME2', y='WTKG3', data=data, whis=10, palette = 'plasma')
plt.yscale('log')

plt.xlabel('Income Group')
plt.ylabel('Weight (kg)')
plt.title('Weight vs. Income');
```

The interquartile ranges were generally consistent across all income groups, suggesting that the middle 50% of individuals in each group have a similar range of heights.

However, there was a slight increase in the mean height within the higher-income groups compared to lower-income groups. This may indicate that individuals with greater income, on average, tend to be taller.

Additionally, it's worth noting that groups 1 and 6 have a wide range of heights, suggesting a presence of outliers that may skew the data. This variability raises questions about the underlying factors that contribute to height disparities within these specific socioeconomic groups, and warrants further investigation.

## Correlation


### Importing packages  
We import the packages `numpy`, `pandas`, and `matplotlib.pyplot` to use to show an example where there is a clear relationship between the variables, but the quantified correlation is low.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Generating fake data for y=cos(x)  
For the x values, represented by variable `x`, we create 100 data points in the range -$\pi$ to $\pi$ using `np.linspace()`. For the y values, represented by variable `y`, we calculate `y` by using the x values and the function `np.cos(x)`. We jitter `y` by adding random noise using `np.random.normal()` which adds noise from a random normal distribution with mean of 0 and standard deviation of 0.05.

```
x = np.linspace(-np.pi, np.pi, 100)
y = np.cos(x) + np.random.normal(0, 0.05, len(x))
```

### Visual representation of y=cos(x) relationship  
Using `plt.plot()` the x and y values are plotted on a scatter plot with the parameter `alpha=0.5` to be able to see if there are any overlaps among the data points.

```
plt.plot(x, y, 'o', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot of y=cos(x) from -π to π');
```

We can see a relationship between `x` and `y` from the scatter plot.

### Looking at the correlation coefficient
We are able to obtain the correlation coefficient between the variables `x` and `y` by using the numpy function `corrcoef()` which exists in the first row and second column of the correlation matrix created by the function (the first row, first column and the second row, second column will be a correlation with itself giving a correlation of 1).

```
np.corrcoef(x, y)[0][1]
```

The correlation coefficient tells us there is a very weak correlation between `x` and `y`.

### Explanation for the low correlation coefficient  
As we can see from the scatter plot, there is a relationship between `x` and `y` for y=cos(x) in the range -$\pi$ to $\pi$. The relationship we see is nonlinear since as `x` increases, `y` oscillates between -1 and 1. Despite what we see, if we quantify the relationship using the correlation coefficient, the coefficient proposes a very weak correlation between `x` and `y` as it is very close to zero. However, it stands that if we have the equation for the relationship y=cos(x), and we know `x`, we are able to competently predict `y`. This weak correlation given by the correlation coefficient regardless of us being able to see an association visually is due to its nature of only being able to quantify the strength of a linear relationship.

## Simple regression

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
brfss = pd.read_hdf('brfss.hdf5', 'brfss')
```

```
brfss.head()
```

```
brfss.describe()
```

```
income = brfss['INCOME2']
veggie = brfss['_VEGESU1']
```

Because the columns are arranged in discrete form, we will add some random noise to normalize the values. This process is called **jittering**. From the demo, we will keep the same mean and standard deviation of 0 and 2 respectively.

```
noise = np.random.normal(0, 2, size=len(brfss))
income_jitter = income + noise
veggie_jitter = veggie + noise
```

To take into account any null values, we will remove those from our chosen variables prior to running the linear regression.

```
subset = brfss.dropna(subset=['INCOME2','_VEGESU1'])
income_clean = subset['INCOME2']
veggie_clean = subset['_VEGESU1']

res_vi = linregress(veggie_clean, income_clean)
res_vi._asdict()
```

Next, let's create our parameters for plotting the linear regression line.

```
fx = np.array([veggie_clean.min(), veggie_clean.max()])
fy = res_vi.intercept + res_vi.slope * fx
```

Finally, time to plot our variables with the trend line.

```
plt.plot(veggie_jitter, income_jitter, 'o', alpha=0.2, markersize = 2)
plt.plot(fx, fy, '-')


plt.xlim([0,20])
plt.ylim([0,20])
plt.xlabel('Vegetable Consumption')
plt.ylabel('Income')
plt.title('Vegetable Consumption vs. Income')
```

Overall, it seems there is a weak correlation between vegetable consumption and overall income. While it seems that the scatter plot tends to go in a positive direction, it is ultimately spread throughout.
