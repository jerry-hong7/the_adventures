## Introduction & Hypothesis

In this project, we will look at six well-known basketball players---Kobe Bryant, Stephen Curry, Larry Bird, Kevin Garnett, Michael Jordan, and Dwyane Wade---who have won championships over the past few decades. We want to see whether their was a positive association between the prevalence of their first names and their winning of a championship or success in a season. Our hypothesis is that after these players won a championship or experienced a breakout season, their names increased in usage/popularity. When a player has a successful year or wins the championship, their popularity will increase, which may affect the name they choose for their child if they admire a particular player. Thus, we expect their names to also grow in popularity. 

This hypothesis is not unfounded. For instance, research from Childcare.co.uk showed a 112% increase in the number of children named after celebrities and film/tv characters between 2009 to 2019. Names such as Ariana, the name of the top-selling female pop star Ariana Grande, and Jesse, a character from the immensely popular show Breaking Bad, were two of the most popular. Richard Conway, the founder of Childrecare.co.UK commented on the trend as well, claiming that '"celebrities have a huge influence on our culture"' and that '"people have historically named their children after their icons."

## Results & Conclusion

While we can't establish causation from this data, there are a number of distinct trends that emerged with each name in the sample. Larry Bird, for instance, won three champions in 1981, 1984 & 1986. While there wasn't a surge in the popularity of the name after this 5 year period, it saw a relatively high popularity in 1988. The name Larry remained consistently popular after this year. We also observed that, sometimes, a name's popularity surges years after the corresponding NBA championship. For example, Kobe won an NBA championship in 2010 (n = 38). In 2021, his name significantly rose in popularity (n = 44). It would be interesting to further investigate this trend by analyzing the average age of NBA championship viewers (e.g. the 2010 championship viewers might've had children in 2021). Lastly, Dwayne Wade won three NBA championships: 2006, 2012, & 2013. His name's popularity remained relatively stable throughout these years (n = 22, n = 23, n = 22, respectively); however, it declined in popularity in the years following, such as 2020 and 2021 (n = 13, n = 15). This suggests that a name might become less popular over time if an NBA player doesn't keep winning championships. Additionally, if the player does not perform on par with their breakout season, we could expect a drop off in popularity.
