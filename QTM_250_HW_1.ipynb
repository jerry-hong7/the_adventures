## **Introduction**

>The **<font color='green'>Dobbs Commmon Table (DCT)</font>** feeds many of Emory's undergraduate students, particularly underclassmen. Located inside the Emory Student Center, it is in close proximity to many residential and academic buildings. The DCT is open 7:30am-10pm on weekdays and 10am-8pm on the weekends. When going at peak times, the crowds are comparable to a New York City subway station during rush hour. The objective of this experiment is to track the number of students entering the DCT during lunch hours, which we will define as 11:30am-1:30pm. 



>Before any predictions are made or data is collected, it is important to explain the motivation behind this particular experiment. The true catalyst behind the decision to research DCT customer fluctuations was, in all honesty, frustration. Each group member observed certain times at the dining hall when their time had been wasted standing in line for food or waiting to swipe into the cafeteria. We believe that, by knowing when the most students are arriving at the dining hall, we can adjust accordingly to optimize personal time management. Furthermore, this data might be useful to DCT workers who feel that the status quo is untenable during peak hours and must be improved by increasing staff. Lastly, we selected DCT instead of Cox Dining Hall or other campus dining services because of its affordability. First and second year students are required to pay for a meal plan. The most economic meal plan options are those that maximize DCT swipes annd minimize the amount of Dooley (dining) dollars. Thus, the DCT is typically the most popular dining attraction for Emory students. 

---

>1. **Thought process**: Variation in the amount of people at the DCT every 15 minute interval between the times of 11:30am and 1:30pm.

>2. **Null hypotheis**: The amount of people entering the DCT betwen the times of 11:30am and 1:30pm will be equal (constant throughout time range)

---

## **Hypothesis**

>We expect that there would be the highest number of students entering the DCT between the times of 12:45pm and 1pm, as the 11:30am classes end at 12:45pm, so students will probably eat after classes finish.

## **Methods**

>We decided to sit outside the entrance of DCT and count the number of students that entered the dining hall between 11:30am to 1:30pm. These hours were chosen because the time range before and after is assumed to have fewer people eating. One member of the group recorded time intervals while two members separately conducted a census. After the test, both members compared the count to ensure accurate results.
