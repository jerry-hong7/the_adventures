# Probability Exercises

Find the probability that a normal random variable with a mean 10 and a standard deviation of 7
* (i) lies between 9 and 12
* (ii) is greater than 25
* (iii) is less than 13

```python
# Use print() for each command to show your work.
print(stats.norm.cdf(12,10,7)-stats.norm.cdf(9,10,7))
print(1-stats.norm.cdf(25,10,7))
print(stats.norm.cdf(15,10,7))
```

# Practice with Random Number Generation and Histograms

* Set the seed at 245
* Generate a vector y of 100 observations from the normal distribution with mean=3 and standard deviation sd=7 
* Generate a vector z of 100 observations from the normal distribution with mean=30 and sd=3
* Generate a vector u of 10,000 observations from the uniform distribution 

```python
# Set seed
np.random.seed(245)


# Generate vector y
y = stats.norm.rvs(size=100, loc=3, scale=7)


# Generate vector z
z= stats.norm.rvs(size=100, loc=30, scale=3)


# Generate vector u
u = stats.uniform.rvs(size=10000)

```

* Create a histogram for y. Change the title to "Histogram of Y" and y-axis label to "Frequency".
* Create a histogram for y and z in the same graph. Add a density curve (smooth curve). Change the title to "Histogram of Y and Z" and y-axis label to "Frequency".
* Create a histogram for u. Add the argument 'bins=20' into the function. Change the title to "Histogram of U" and y-axis label to "Frequency".

```python
# Histogram of Y
y_hist = sns.histplot(y);
y_hist.set(title="Histogram of Y",
          ylabel="Frequency");
```

```python
# Histogram of Y and Z on the same graph
data = pd.DataFrame({'Y': y, 'Z':z})
hist_comp = sns.histplot(data, kde=True);
hist_comp.set(title='Histogram of Y and Z',
             ylabel='Frequency');
```

```python
# Histogram of U
u_hist = sns.histplot(u, bins=20);
u_hist.set(title="Histogram of U",
          ylabel="Frequency");
```
